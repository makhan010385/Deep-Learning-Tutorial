```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Dataset/placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(df['cgpa'],df['resume_score'],hue=df['placed'])
#sns.scatterplot(df, x="cgpa", y="resume_score", hue="placed")
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

Perceptron.png

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
